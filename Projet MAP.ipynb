### Question T1 :


Soit $F$ une primitive double de $f$ sur l'intervalle $[0,1]$  (ie $F'' = f$)


On suppose que $F(0) = F(1) = 0$, quitte à remplacer $F$ par 
$x \rightarrow F(x) - F(0)(1-x) - F(1) x$

Posons $v = u - F$, 

\begin{equation*}
u \hbox{ est solution ssi }
\begin{cases} v'' =  0 \\ v(0) =  a \\  v(1) =  b \end{cases}
\hbox{ ie }
\begin{cases}
v = A(1-x) + Bx \\ A = a \\ B = b
\end{cases}
\end{equation*}

Le problème admet donc une unique solution : 

$$u : x \rightarrow F(x) + a(1-x) + bx $$

\vspace{5mm}

**Cas 1 :** $f(x) = sin(2 \pi x)$ , $a=b=0$

\begin{equation*}
u(x) = - \frac{sin(2 \pi x)}{4 \pi ^2}
\end{equation*}

Cette solution convient et est donc unique

**Cas 2 :** $f(x) = 0$ , $a,b \in \mathbb{R}$

$F(x) = 0$ est bien une primitive double de f sur $[0,1]$ telle que $F(0) = F(1) = 1$

\begin{equation*}
u(x) = a(1-x) + bx
\end{equation*}

### Question T3 :

Soit $x \in I_h$

\begin{equation*}
\{ \zeta_{M+1} = \dots = \zeta_{2M} = 1 \} \subset \{ T^x \leq 2M \}
\end{equation*}


En effet, soit $\omega \in \Omega \hbox{ tel que } \zeta_{M+1}(\omega) = \dots = \zeta_{2M}(\omega) = 1$

Si $X_M(\omega) \notin I$ alors clairement $T^x(\omega) \leq 2M$

Si $X_M(\omega) \in I$ alors $X_{2M}(\omega) = X_M(\omega) + \sum_{n=M+1}^{2M} h = X_M(\omega) + 1 \geq 1$ donc $T^x(\omega) \leq 2M$
\vspace{5mm}

\begin{equation*}
\mathbb{P}( T^x \leq 2M | T^x > M) \geq \mathbb{P}(\zeta_{M+1} = \dots = \zeta_{2M} = 1 | T^x > M)
\end{equation*}

Or $\{ T^x > M \}$ s'écrit en fonction uniquement des $(\zeta_i)_{1\leq i \leq M}$ donc on en conclut par indépendance des $(\zeta_i)$ que $\mathbb{P}(\zeta_{M+1} = \dots = \zeta_{2M} = 1 | T^x > M) = \mathbb{P}(\zeta_{M+1} = \dots = \zeta_{2M} = 1)$

\begin{equation*}
\mathbb{P}( T^x \leq 2M | T^x > M) \geq \mathbb{P}(\zeta_{M+1} = \dots = \zeta_{2M} = 1) = \left( \frac{1}{2} \right)^M
\end{equation*}

Donc finalement:

$$
\mathbb{P}( T^x > 2M | T^x > M) \leq 1 - \left( \frac{1}{2} \right)^M
$$
\vspace{2mm}

En fait, on a même plus généralement avec ce même raisonnement (en décallant les indices) que :

\begin{equation*}
\forall k \in \mathbb{N}, \ \mathbb{P}( T^x > (k+1)M \ | \ T^x > kM) \leq 1 - \left( \frac{1}{2} \right)^M
\end{equation*}

On remarque que cette expression est bien valable pour $k=0$ car $x \notin \{0,1\}$ en adaptant légèrement la démonstation de $(3)$.

\vspace{2mm}

On obtiens alors rapidement par récurrence sur $k \in \mathbb{N}$
$$
\forall k \in \mathbb{N} , \ \mathbb{P}(T^x > kM) \leq \left[ 1 - \left( \frac{1}{2} \right)^M \right]^k
$$

\vspace{5mm}

Démonstrons que pour $Y \geq 0$ v.a.r à valeurs dans $\mathbb{N}$

\begin{equation}
\sum_{n=0}^{+\infty} \mathbb{P}(Y > n) < +\infty \implies \mathbb{E}(Y) < +\infty
\end{equation}


\vspace{2mm}
\begin{proof}

Supposons le membre de gauche vérifié par Y 

\begin{align*}
\sum_{n=0}^K n\mathbb{P}(Y=n) 
& = \sum_{n=1}^K n (\mathbb{P}(Y>n-1) - \mathbb{P}(Y>n) \\
& = \sum_{n=0}^{K-1} (n+1) \mathbb{P}(Y>n) - \sum_{n=0}^K n\mathbb{P}(Y>n) \\
& = \sum_{n=0}^{K-1} \mathbb{P}(Y>n) - K \mathbb{P}(Y>K)
\end{align*}

Ainsi
\begin{equation*}
\forall K, \sum_{n=0}^K n\mathbb{P}(Y=n) \leq \sum_{n=0}^{K-1} \mathbb{P}(Y>n) \leq \sum_{n=0}^{+\infty} \mathbb{P}(Y>n) < +\infty
\end{equation*}

D'où $\mathbb{E}(Y) < + \infty$

\end{proof}

Utilisons ce résultat pour montrer que $T^x$ est intégrable

\begin{align*}
\sum_{n=0}^{+\infty} \mathbb{P}(T^x > n) & = \sum_{k=0}^{+\infty} \sum_{n=kM}^{(k+1)M-1}\mathbb{P}(T^x > n) \\
                                         & \leq \sum_{k=0}^{+\infty} \sum_{n=kM}^{(k+1)M-1}\mathbb{P}(T^x > kM) \\
                                         & \leq \sum_{k=0}^{+\infty} M \left[ 1 - \left( \frac{1}{2} \right)^M \right]^k \\
                                         & < +\infty
\end{align*}

D'après $(5)$, on en déduit que $T^x \in \mathbb{L}^1$

### Question S1 :

Voici tout d'abord une version naïve de l'algorithme

```python
def marcheNaive(x, h = h_glob, out = "ret"):
    '''Simule une marche sur I en partant de x'''
    #Si out == "plot", affiche le graphe, si out == ret, renvoie les Xn
    #Sinon ne fait rien
    X = [x]
    t = 0
    while 0<X[t]<1:
        gamma = 2*npr.binomial(1,1/2) -1 #Simulation d'une variable gamma
        X.append(X[t] + h*gamma)
        t += 1
    if out == "plot":
        plt.plot(X)
        plt.show()
    elif out == "ret":
        return X
    

def marcheNaive5(x = 1/2, h = h_glob):
    plt.plot(marcheNaive(x,h = h))
    plt.plot(marcheNaive(x,h = h))
    plt.plot(marcheNaive(x,h = h))
    plt.plot(marcheNaive(x,h = h))
    plt.plot(marcheNaive(x,h = h))
    plt.show()

marcheNaive5()
```

Cette version repose sur la boucle while de Python, que l'on sait lente comparée aux opérations de Numpy. Nous proposons donc une seconde version du programme utilisant plus efficacement la bibliothèque Numpy et avec moins de passage dans la boucle while.

```python
def marche(x, h = h_glob, out = "ret"):
    '''Simule une marche sur I en partant de x'''
    M = int(1/h)
    #On réalise des tirages de N variables gamma simultanées
    N = 10*M
    position = x
    X = np.array([x])
    while True:
        #On simule les variables gamma par paquet de N
        gamma = 2*npr.binomial(size=N, n=1, p=1/2) -1
        
        #Xc contient les M X_n simulées
        #On ajoute position qui correspond au X_n final
        #du précédent passage dans la boucle while
        Xc = h*np.cumsum(gamma) + position
        
        #Recherche des n pour lesquels la marche sort de I
        critical_index = np.nonzero((0>=Xc) | (Xc>=1))[0]
        if critical_index.size != 0 :
            X = np.concatenate((X,Xc[:critical_index[0]+1]))
            break
        position = Xc[N-1]
        X = np.concatenate((X,Xc))
    if out == "plot":
        plt.plot(X)
        plt.show()
    elif out == "ret":
        return X

def marche5(x = 1/2, h = h_glob):
    plt.plot(marche(x,h = h))
    plt.plot(marche(x,h = h))
    plt.plot(marche(x,h = h))
    plt.plot(marche(x,h = h))
    plt.plot(marche(x,h = h))
    plt.show()

        
marche5()
```

On peut comparer ces deux implémentation de la marche aléatoire :

```python
vmarcheNaive = np.vectorize(marcheNaive)
vmarche = np.vectorize(marche)

def compareMarches(N):
    X = 1/2 + np.zeros(N)
    tnaif = time.process_time()
    vmarcheNaive(X, out = "")
    tnaif = time.process_time() - tnaif
    t = time.process_time()
    vmarche(X, out = "")
    t = time.process_time() - t
    return (tnaif,t)

compareMarches(10000)
```

En expérimentant avec la fonction compareMarches pour N = 100000, on remarque que la fonction moins naîve marche est en moyenne 10 fois plus efficace que marcheNaive

```python
def X_Tx(x, h = h_glob):
    '''Renvoie la valeur de X_(Tx) pour une marche simulée'''
    M = int(1/h)
    N_tir = 10*M
    position = x
    while True:
        gamma = 2*npr.binomial(size=N_tir, n=1, p=1/2) -1
        Xc = h*np.cumsum(gamma) + position
        critical_index = np.nonzero((0>=Xc) | (Xc>=1))[0]
        if critical_index.size != 0:
            if Xc[critical_index[0]] <= 0:
                return 0
            return 1
        position = Xc[N_tir-1]

vect_X_Tx = np.vectorize(X_Tx)
```

```python
a_glob = 1
b_glob = 2

def v(x, N, a = a_glob, b = b_glob, h = h_glob):
    '''Calcule la valeur de v au point x avec N points'''
    sample = vect_X_Tx(x + np.zeros(N),h)
    #Simulation de indicatrice de (X_Tx = 1)
    esp = np.mean(sample)
    return a + (b-a)*esp

vect_v = np.vectorize(v)


def trace_v(nb_points_affiche, N, a = a_glob, b = b_glob, h = h_glob):
    '''Trace la fonction v calculée avec N points et la solution réelle'''
    X = np.linspace(0,1,nb_points_affiche)
    Y_exp = vect_v(X,N,a,b,h)
    Y_th = a + (b-a)*X
    plt.plot(X,Y_exp)
    plt.plot(X,Y_th)
    plt.show()

trace_v(10,4000)
```

```python
def erreur_h(N, a = a_glob, b = b_glob):
    plt.xscale("log")
    plt.yscale("log")
    H = 10**(-np.linspace(0,3,20))
    V = vect_v(1/2, N , a,b, h = H)
    Delta = np.abs(V - (a+b)/2)
    plt.plot(H,Delta)
    plt.show()

erreur_h(400)
```
